### Financial Connection

Markets move.
Prices change.
Returns are calculated mathematically.

Let us compute a simple return.

```python
price_yesterday = 100
price_today = 105

daily_return = (price_today - price_yesterday) / price_yesterday

print("Daily Return:", daily_return)
```

# Sprint 2 - Working with Time Series

Financial data is sequential.
We store price history inside lists.

```python
# A simple price series
prices = [100, 102, 101, 105]

print("First price:", prices[0])
print("Number of prices:", len(prices))
```

## Looping Through Time

A loop allows us to move through time step by step.
We compute returns for each day.

```python
returns = []

for i in range(1, len(prices)):
    r = (prices[i] - prices[i - 1]) / prices[i - 1]
    returns.append(r)

print("Returns:", returns)
```

### Quant Connection

This is how early financial systems were built.

But loops become slow with large datasets.

We now upgrade our thinking.

# Sprint 3 - Vectorized Computation

Professional quantitative systems do not loop manually.

They operate on entire arrays at once.

This is called vectorization.

```python
import numpy as np

prices_array = np.array([100, 102, 101, 105])

returns_vectorized = (prices_array[1:] - prices_array[:-1]) / prices_array[:-1]

print("Vectorized Returns:", returns_vectorized)
```

## What Just Happened?

prices_array[1:]  → shifts forward  
prices_array[:-1] → shifts backward  

Subtraction happens element-wise.

This is:
- Faster
- Cleaner
- Scalable

This is how quants compute.

# Sprint 4 - DataFrames

Markets generate structured datasets.
We use Pandas to manage them.

```python
import pandas as pd

data = pd.DataFrame({
    "Price": [100, 102, 101, 105, 110]
})

data
```

## Computing Returns Automatically

```python
data["Returns"] = data["Price"].pct_change()

data
```

## Rolling Volatility

Markets are not stable.
Risk changes over time.

A rolling window measures time-varying volatility.

```python
data["Volatility"] = data["Returns"].rolling(3).std()

data
```

# Sprint 5 - Execute

You now combine everything.

Build a mini quant workflow:
- Price series
- Returns
- Rolling volatility

```python
import numpy as np
import pandas as pd

# Simulated price data
prices = np.array([100, 101, 103, 98, 105, 110, 108])

# Create DataFrame
df = pd.DataFrame({"Price": prices})

# Compute returns
df["Returns"] = df["Price"].pct_change()

# Compute rolling volatility (3-day window)
df["Volatility"] = df["Returns"].rolling(3).std()

df
```
